```python
def preprocessing(image_path):
    target_width = 28
    target_height = 28
    colour = 'grayscale'
    img =image.load_img(image_path, target_size=(target_height,target_width), color_mode= colour)
    img = image.img_to_array(img)
    img = 255-img
    img = img/255
    img_batch = np.expand_dims(img, axis=0)
    return img_batch
```

```python
image_path = 'digit2.jpg'
img_batch =preprocessing(image_path)
prediction=loaded_model.predict(img_batch) 
np.argmax(prediction[0])
```
